# Reduce Data Set
The original data set had over 51 million rows, so we opted to reduce it by using only the last 6 years (2012-2017) and flattening the rows to contain data for each day as opposed to each hour.

```python
import pandas as pd
import numpy as np

dlst = pd.read_stata("../spring forward/data/stats19/final/final.dta")
```

```python
dlst2 = dlst[dlst["year"] >= 2012]
```

```python
dlst3 = dlst2.groupby(by=["mlong", "mlat", "date", "sunrise6_time", "sunset6_time", "year", "doy", "dow", "longlat", "darkh"], as_index=False).agg(
    {'all_accid': "sum",
     'u25_accid': "sum",
     'o25u45_accid': "sum",
     'o45u65_accid': "sum",
     'o65_accid': "sum",
     'ped_accid': "sum",
     'ped_u45_accid': "sum",
     'ped_o45_accid': "sum",
     'fat_accid': "sum",
     'ser_accid': "sum",
     'slight_accid': "sum"}
)
```
